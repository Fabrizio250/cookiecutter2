# Subject analysis
In this notebbok we will analyze the population of OASIS3 dataset

```python
import pandas as pd
import seaborn as sns
from os.path import join

subjects = pd.read_csv("../data/raw/subjects.csv")
print(f"Read subject file, it contains {len(subjects)} subjects")

PLOT_PATH = "../reports/figures/"
```

## Population overview
We analyze the subjects in the dataset to see how are distributed according to:
- Dominant Hand
- Gender
- Ethnicity

```python
g = sns.countplot(data=subjects, x='Gender')
g.set_title("Subject distribution by gender")
g.figure.savefig(join(PLOT_PATH,"subjects_by_gender.png"), bbox_inches='tight', dpi=300)
```

```python
g = sns.countplot(data=subjects, x='Hand')
g.set_title("Subject distribution by dominant hand")
g.figure.savefig(join(PLOT_PATH,"subjects_by_hand.png"), bbox_inches='tight', dpi=300)
```

```python
g = sns.countplot(data=subjects, x='Race')
g.set_title("Subject distribution by race")
g.figure.savefig(join(PLOT_PATH,"subjects_by_race.png"), bbox_inches='tight', dpi=300)
```

```python
g = sns.countplot(data=subjects, x='Race', hue="Gender")
g.set_title("Subject distribution by race")
g.figure.savefig(join(PLOT_PATH,"subjects_by_race_and_gender.png"), bbox_inches='tight', dpi=300)
```

## Conclusions
From the charts above we can observe a predominance of caucasian people in the dataset, we cannot say if this is because of the low susceptibility to AD of other ethnicities or other causes.
As expected the most common dominant hand is the right one.
From what concern the gender the dataset is almost equally distributed with slightly more female rather than male subjects.
